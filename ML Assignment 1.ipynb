***Q6 : Train your model on different train test ratio and train the models and note down there accuracies***

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Load data from a CSV file
data = pd.read_csv("C:/Users/SABNAM NAYAK/Downloads/FuelConsumption (10).csv")

# Split the data into training and testing sets with different ratios
ratios = [0.5, 0.6, 0.7, 0.8]
for ratio in ratios:
    msk = np.random.rand(len(data)) < ratio
    train = data[msk]
    test = data[~msk]

    # Train the model on the training data
    regr = LinearRegression()
    train_x = np.asanyarray(train[['FUELCONSUMPTION_COMB']])
    train_y = np.asanyarray(train[['CO2EMISSIONS']])
    regr.fit(train_x, train_y)

    # Test the model on the testing data and calculate the accuracy
    test_x = np.asanyarray(test[['FUELCONSUMPTION_COMB']])
    test_y = np.asanyarray(test[['CO2EMISSIONS']])
    test_y_hat = regr.predict(test_x)
    accuracy = r2_score(test_y_hat, test_y)
    print("Train-test ratio:", ratio, "Accuracy:", accuracy)
```

**WAY 1: linear regression on 5 different variables**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load the dataset
data = pd.read_csv("C:/Users/SABNAM NAYAK/Downloads/houseRegressionStudents (1) (1).csv")
data.head()

# Split the data into train and test sets
msk = np.random.rand(len(data)) < 0.80
train = data[msk]
test = data[~msk]

# Create a list of features to iterate through
features = ['MSSubClass', 'LotArea', 'EnclosedPorch', 'MoSold', 'YrSold']

# Iterate through each feature and train a linear regression model
for feature in features:
    regr = LinearRegression()
    train_x = np.asanyarray(train[[feature]])
    train_y = np.asanyarray(train[['SalePrice']])
    regr.fit(train_x, train_y)
    
    # Print the coefficients and intercept
    print('Feature:', feature)
    print('Coefficients:', regr.coef_)
    print('Intercept:', regr.intercept_)
    
    # Plot the training data and the linear regression line
    plt.scatter(train[feature], train.SalePrice, color='pink')
    plt.plot(train_x, regr.predict(train_x), c='blue')
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.show()

    # Evaluate the model on the test set
    test_x = np.asanyarray(test[[feature]])
    test_y = np.asanyarray(test[['SalePrice']])
    print('Residual sum of squares: %.2f' % np.mean((regr.predict(test_x) - test_y) ** 2))
    print('R2 score: %.2f' % regr.score(test_x, test_y))
```
