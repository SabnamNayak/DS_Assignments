```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar=StandardScaler()
```

```python
scaled_values=scalar.fit_transform(df)
```

```python
scaled_values
```

```python
scaled_df=pd.DataFrame(scaled_values,columns=iris.feature_names)
```

```python
scaled_df
```

```python
scaled_df.info()
```

```python
scaled_df.isnull().sum()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(scaled_df)
```

```python
principalComponents
```

```python
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['PC1', 'PC2'])
```

```python
principalDf
```

```python
principalDf['Target']=iris.target
```

```python
principalDf
```

```python
plt.figure(figsize=(12,8))
colormap=np.array(['red','blue','black'])
plt.scatter(principalDf.PC1,principalDf.PC2,c=colormap[principalDf.Target],s=50)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("Principal Component axis")
```

```python
pca.explained_variance_ratio_
```

```python
plt.figure(figsize=(12,8))
plt.bar(['PC1','PC2'],pca.explained_variance_ratio_)
```

```python
pca.explained_variance_ratio_.sum()
```
